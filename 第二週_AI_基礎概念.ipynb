- 機器學習的基本概念：線性迴歸

  - 這裡可以使用一個簡單的線性迴歸示例，來介紹機器學習的核心思想 https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html


- 小任務：

  - 請隨意生成數據，再重新觀察畫出來的結果，並與你所曾經學過的二元N次方程式公式解進行比較
  - https://www.liveism.com/live-concept.php?q=%E5%B7%B2%E7%9F%A5%E7%9B%B4%E7%B7%9A%E4%B8%8A%E5%85%A9%E9%BB%9E%E6%B1%82%E7%9B%B4%E7%B7%9A%E6%96%B9%E7%A8%8B%E5%BC%8F

```python
import numpy as np

# 給定的數據
X = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])

# 計算 x 和 y 的均值
x_mean = np.mean(X)
y_mean = np.mean(y)

# 計算 m（斜率）
m = np.sum((X - x_mean) * (y - y_mean)) / np.sum((X - x_mean)**2)

# 計算 b（截距）
b = y_mean - m * x_mean

# 顯示結果
print(f"斜率 (m): {m}")
print(f"截距 (b): {b}")

# 使用封閉解進行預測
def predict(x):
    return m * x + b

# 測試預測
predictions = predict(X)
print(f"預測結果: {predictions}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 生成數據
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1, 2, 3, 4, 5])

# 創建線性迴歸模型並訓練
model = LinearRegression()
model.fit(X, y)

# 預測
y_pred = model.predict(X)

# 繪製圖表
plt.scatter(X, y, color='blue')  # 原始數據點
plt.plot(X, y_pred, color='red')  # 回歸線
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.show()

# 增加新的數據點進行預測
NewX = np.array([9, 10, 11, 12, 13]).reshape(-1, 1)
NewY = model.predict(NewX)

# 繪製新數據的預測結果
plt.scatter(X, y, color='blue')  # 原始數據點
plt.scatter(NewX, NewY, color='green')  # 新數據點
plt.plot(X, y_pred, color='red')  # 回歸線
plt.title('Linear Regression with New Data')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

深度學習的簡單範例：使用 TensorFlow/Keras 構建神經網絡

使用一個非常簡單的神經網絡來展示深度學習的基礎概念，這裡以經典的 MNIST 數字識別為例。

- 小任務：

  - 透過 http://playground.tensorflow.org/#activation=tanh&batchSize=10&dataset=circle&regDataset=reg-plane&learningRate=0.03&regularizationRate=0&noise=0&networkShape=4,2&seed=0.92179&showTestData=false&discretize=false&percTrainData=50&x=true&y=true&xTimesY=false&xSquared=false&ySquared=false&cosX=false&sinX=false&cosY=false&sinY=false&collectStats=false&problem=classification&initZero=false&hideText=false 感受深度學習是如何運作的

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

# 加載 MNIST 數據
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# 視覺化部分：顯示前 10 張手寫數字圖片
def visualize_mnist_data(images, labels):
    plt.figure(figsize=(10, 1))
    for i in range(10):
        plt.subplot(1, 10, i+1)
        plt.imshow(images[i], cmap='gray')
        plt.axis('off')
        plt.title(labels[i])
    plt.show()

# 顯示訓練集中前 10 張圖像及其標籤
visualize_mnist_data(train_images, train_labels)

# 數據預處理
train_images = train_images.reshape((60000, 28, 28, 1)).astype('float32') / 255
test_images = test_images.reshape((10000, 28, 28, 1)).astype('float32') / 255

# 構建卷積神經網絡模型
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# 添加全連接層
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

# 編譯模型
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 訓練模型（由於時間限制，這裡只訓練一個 epoch）
model.fit(train_images, train_labels, epochs=1, batch_size=64)

# 測試模型性能
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'測試準確率: {test_acc}')

# 視覺化測試集中前 10 張圖像及其標籤
visualize_mnist_data(test_images[:10].reshape(10, 28, 28), test_labels[:10])
```
