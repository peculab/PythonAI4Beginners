- 電腦視覺對應到人的眼睛：
  - 電腦視覺（Computer Vision）是讓機器學習從圖像和視覺訊息中提取有用數據的技術。就像人的眼睛感知外部世界並向大腦傳遞視覺訊息一樣，電腦視覺能夠識別物體、檢測圖像中的模式，並做出反應。應用場景包括臉部識別、自動駕駛等。

- 自然語言處理對應到人的語言理解和表達能力：
  - 自然語言處理（Natural Language Processing, NLP）讓 AI 能夠理解、生成和回應人類語言。這可以與人類的大腦處理語言的能力相對應，包括聽、說、讀、寫。NLP 的應用範圍包括語音助手、聊天機器人、翻譯系統等。

- 強化學習對應到人的學習與決策能力：
  - 強化學習（Reinforcement Learning）是 AI 模仿人類學習和決策過程的一種技術。它通過試錯和獎勵機制來改進行為，這類似於人類學習如何在不同情境下做出最佳決策。

- 語音識別對應到人的聽覺：
  - 語音識別技術讓 AI 能夠像人類的耳朵一樣理解語音訊息，並將語音轉換為文字。語音助手（如 Siri 和 Alexa）就是語音識別技術的應用。

- 機器學習對應到人的推理和模式識別能力：
  - 機器學習（Machine Learning）是 AI 模擬人類從數據中學習和推理的能力。通過大量數據訓練，機器學習能夠識別模式並進行預測，就像人類通過經驗和學習來形成知識和做出判斷一樣。

# AI 的基本概念與歷史

人工智慧（AI）是一個讓機器具備模擬人類智能的能力的領域，涵蓋自動化、機器學習、深度學習等技術

其歷史可以追溯到 1950 年代的圖靈測試

- 1956 年：AI 的概念正式提出
- 1970-1980 年代：專家系統興起
- 2000 年代以後：機器學習和深度學習技術迅速發展，AI 開始應用於多個領域

- 機器學習的基本概念：線性迴歸

```python
import numpy as np

# 給定的數據
X = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])

# 計算 x 和 y 的均值
x_mean = np.mean(X)
y_mean = np.mean(y)

# 計算 m（斜率）
m = np.sum((X - x_mean) * (y - y_mean)) / np.sum((X - x_mean)**2)

# 計算 b（截距）
b = y_mean - m * x_mean

# 顯示結果
print(f"斜率 (m): {m}")
print(f"截距 (b): {b}")

# 使用封閉解進行預測
def predict(x):
    return m * x + b

# 測試預測
predictions = predict(X)
print(f"預測結果: {predictions}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 生成數據
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1, 2, 3, 4, 5])

# 創建線性迴歸模型並訓練
model = LinearRegression()
model.fit(X, y)

# 預測
y_pred = model.predict(X)

# 繪製圖表
plt.scatter(X, y, color='blue')  # 原始數據點
plt.plot(X, y_pred, color='red')  # 回歸線
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.show()

# 增加新的數據點進行預測
NewX = np.array([9, 10, 11, 12, 13]).reshape(-1, 1)
NewY = model.predict(NewX)

# 繪製新數據的預測結果
plt.scatter(X, y, color='blue')  # 原始數據點
plt.scatter(NewX, NewY, color='green')  # 新數據點
plt.plot(X, y_pred, color='red')  # 回歸線
plt.title('Linear Regression with New Data')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

深度學習的簡單範例：使用 TensorFlow/Keras 構建神經網絡

使用一個非常簡單的神經網絡來展示深度學習的基礎概念，這裡以經典的 MNIST 數字識別為例。

- 小任務：

  - 透過 http://playground.tensorflow.org/#activation=tanh&batchSize=10&dataset=circle&regDataset=reg-plane&learningRate=0.03&regularizationRate=0&noise=0&networkShape=4,2&seed=0.92179&showTestData=false&discretize=false&percTrainData=50&x=true&y=true&xTimesY=false&xSquared=false&ySquared=false&cosX=false&sinX=false&cosY=false&sinY=false&collectStats=false&problem=classification&initZero=false&hideText=false 感受深度學習是如何運作的

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

# 加載 MNIST 數據
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# 視覺化部分：顯示前 10 張手寫數字圖片
def visualize_mnist_data(images, labels):
    plt.figure(figsize=(10, 1))
    for i in range(10):
        plt.subplot(1, 10, i+1)
        plt.imshow(images[i], cmap='gray')
        plt.axis('off')
        plt.title(labels[i])
    plt.show()

# 顯示訓練集中前 10 張圖像及其標籤
visualize_mnist_data(train_images, train_labels)

# 數據預處理
train_images = train_images.reshape((60000, 28, 28, 1)).astype('float32') / 255
test_images = test_images.reshape((10000, 28, 28, 1)).astype('float32') / 255

# 構建卷積神經網絡模型
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# 添加全連接層
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

# 編譯模型
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 訓練模型（由於時間限制，這裡只訓練一個 epoch）
model.fit(train_images, train_labels, epochs=1, batch_size=64)

# 測試模型性能
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'測試準確率: {test_acc}')

# 視覺化測試集中前 10 張圖像及其標籤
visualize_mnist_data(test_images[:10].reshape(10, 28, 28), test_labels[:10])
```
