```python
# 從 gsheets 的 All-whiteboard-device 載入 sheets
data_111 = gsheets.worksheet('111學習成效').get_all_values()
data_112 = gsheets.worksheet('112學習成效').get_all_values()
data_math_1121 = gsheets.worksheet('112-1因材網數學').get_all_values()
data_math_1122 = gsheets.worksheet('112-2因材網數學').get_all_values()

df_111 = pd.DataFrame(data_111[1:], columns=data_111[0])
df_112 = pd.DataFrame(data_112[1:], columns=data_112[0])
math_1121 = pd.DataFrame(data_math_1121[1:], columns=data_math_1121[0])
math_1122 = pd.DataFrame(data_math_1122[1:], columns=data_math_1122[0])
```

```python
subset_dataframe = df_111.iloc[:, :11]
```

```python
subset_dataframe = pd.concat([subset_dataframe, df_112.iloc[:, 1:11]], axis=1)
```

```python
comprehensive_scores = subset_dataframe.applymap(
    lambda x: int(float(x)) if str(x).replace('.', '', 1).isdigit() else x
)
```

```python
comprehensive_scores
```

```python
import plotly.express as px

# Create a subset of data for visualization
subset_data = comprehensive_scores.iloc[:, 1:]  # Exclude the '座號' column
subset_data = subset_data.T  # Transpose for plotting
subset_data.columns = comprehensive_scores['座號']  # Set student IDs as columns

# Create a long-format DataFrame for Plotly
plotly_data = subset_data.reset_index().melt(id_vars='index', var_name='座號', value_name='成績')
plotly_data.rename(columns={'index': '成績類別'}, inplace=True)

# Plot the trend using Plotly
fig = px.line(
    plotly_data,
    x='成績類別',
    y='成績',
    color='座號',
    markers=True,
    title='學生成績變化趨勢',
    labels={'成績類別': '成績類別', '成績': '成績', '座號': '學生座號'},
    template='plotly_white'
)

fig.update_layout(
    xaxis=dict(tickangle=45),
    width=1000,
    height=600
)

# Show the plot
fig.show()
```

```python
# Convert time columns to seconds
def time_to_seconds(time_str):
    try:
        h, m, s = map(int, time_str.split(':'))
        return h * 3600 + m * 60 + s
    except ValueError:
        return 0  # Handle cases where the time format is invalid

# Identify the columns to convert
columns_to_convert = ['影片瀏覽', '練習題測驗', '單元測驗', '縱貫測驗', '動態評量測驗', '使用總時數']

# Apply conversion for each column
for column in columns_to_convert:
    if column in math_1121.columns:
        math_1121[column] = math_1121[column].apply(time_to_seconds)
    if column in math_1122.columns:
        math_1122[column] = math_1122[column].apply(time_to_seconds)
```

```python
math_1121.columns = ['年級', '班級', '座號', '1121任務完成', '1121任務未完成', '1121影片瀏覽', '1121練習題測驗', '1121單元測驗', '1121縱貫測驗', '1121動態評量測驗', '1121使用總時數']
```

```python
math_1122.columns = ['年級', '班級', '座號', '1122任務完成', '1122任務未完成', '1122影片瀏覽', '1122練習題測驗', '1122單元測驗', '1122縱貫測驗', '1122動態評量測驗', '1122使用總時數']
```

```python
allstudents = allstudents.reset_index()
```

```python
results = pd.concat([math_view, allstudents.iloc[:, 1:]], axis=1)
```

```python
results
```

```python
import plotly.express as px

# Prepare the DataFrame for visualization
data_for_plot = results.copy()

# Ensure 'std' and 'mean' are numeric
data_for_plot['std'] = pd.to_numeric(data_for_plot['std'], errors='coerce')
data_for_plot['mean'] = pd.to_numeric(data_for_plot['mean'], errors='coerce')
```

```python
# 繪製散點圖
fig = px.scatter(
    data_for_plot,
    x='1121使用總時數',
    y='mean',
    size='std',
    color='1122使用總時數'
)

fig.update_traces(textposition='top center')
fig.update_layout(template='plotly_white', width=800, height=600)

# 顯示圖表
fig.show()
```
