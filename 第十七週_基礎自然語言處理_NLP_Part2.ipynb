```python
import jieba
import pandas as pd
from collections import Counter
import re
import plotly.express as px
from gensim import corpora, models
from gensim.models import LdaModel
import pyLDAvis.gensim_models
import numpy as np
from sklearn.decomposition import PCA
```

```python
# 只保留中文字
def extract_chinese(text):
    return ''.join(re.findall(r'[\u4e00-\u9fa5]', text))
```

```python
df['Chinese_Content'] = df['內容'].apply(extract_chinese)

# 使用 Jieba 斷詞
df['Tokenized'] = df['Chinese_Content'].apply(lambda x: [word for word in jieba.cut(x) if len(word) > 1])
```

```python
# 1. Create a dictionary and corpus from the tokenized data
dictionary = corpora.Dictionary(df['Tokenized'])
corpus = [dictionary.doc2bow(text) for text in df['Tokenized']]
tfidf_model = models.TfidfModel(corpus)
corpus_tfidf = tfidf_model[corpus]

# 2. Train the LDA model (adjust num_topics based on your needs)
lda_model = LdaModel(corpus_tfidf, num_topics=3, id2word=dictionary, passes=30)

# 3. Show the topics (top 5 words in each topic)
# 初始化一個空列表來儲存每個主題的字串
topics_as_strings = []

# 提取每個主題的內容並將其轉換為字串
for idx, topic in lda_model.print_topics(num_topics=3, num_words=30):
    topic_str = f"Topic {idx}: {topic}"
    topics_as_strings.append(topic_str)

# 現在 topics_as_strings 是包含所有主題的字串列表
# 可以查看或處理這些字串
for topic in topics_as_strings:
    print(topic)  # 或者將其寫入檔案等操作

# 4. Visualize the topics using pyLDAvis (interactive visualization)
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)

# 5. Get topic distribution for each document
topic_distributions = [lda_model.get_document_topics(doc) for doc in corpus]

# Prepare data for dimensionality reduction (PCA)
topic_matrix = np.zeros((len(corpus), lda_model.num_topics))

for i, doc in enumerate(topic_distributions):
    for topic, prob in doc:
        topic_matrix[i, topic] = prob

# 6. Reduce the dimensionality to 2D using PCA
pca = PCA(n_components=3)
reduced_data = pca.fit_transform(topic_matrix)

# 7. Create a DataFrame for Plotly visualization with the titles
df_vis = pd.DataFrame(reduced_data, columns=['PCA1', 'PCA2', 'PCA3'])
df_vis['Document'] = df['標題']  # Add title as label

# 8. Visualize the result with Plotly
fig = px.scatter_3d(df_vis, x='PCA1', y='PCA2', z='PCA3', text='Document', title="LDA Topic Modeling Results",
                 labels={'PCA1': 'PCA Dimension 1', 'PCA2': 'PCA Dimension 2', 'PCA3': 'PCA Dimension 3'})
fig.update_traces(textposition="top center")  # Adjust text position for better visibility
fig.show()
```
